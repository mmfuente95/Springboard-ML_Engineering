### Creating the Tables

In this section, we will be creating the three tables of interest and populate them with the data from the CSV files already available to you.

Once you have done this, please remember to execute the following code to build the dataframes which will be saved as tables in our database

```python
# File location and type
file_location_bookings = "/FileStore/tables/Bookings-4.csv"
file_location_facilities = "/FileStore/tables/Facilities-3.csv"
file_location_members = "/FileStore/tables/Members-3.csv"

file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
bookings_df = (spark.read.format(file_type) 
                    .option("inferSchema", infer_schema) 
                    .option("header", first_row_is_header) 
                    .option("sep", delimiter) 
                    .load(file_location_bookings))

facilities_df = (spark.read.format(file_type) 
                      .option("inferSchema", infer_schema) 
                      .option("header", first_row_is_header) 
                      .option("sep", delimiter) 
                      .load(file_location_facilities))

members_df = (spark.read.format(file_type) 
                      .option("inferSchema", infer_schema) 
                      .option("header", first_row_is_header) 
                      .option("sep", delimiter) 
                      .load(file_location_members))
```

#### Q6: You'd like to get the first and last name of the last member(s) who signed up. Do not use the LIMIT clause for your solution.

```python
%sql
WITH order_members AS (
select firstname, surname, joindate, RANK() OVER (ORDER BY joindate DESC) as ordered_signups
from members)

SELECT firstname, surname
FROM order_members
WHERE ordered_signups = 1


```

####  Q7: How can you produce a list of all members who have used a tennis court?
- Include in your output the name of the court, and the name of the member formatted as a single column. 
- Ensure no duplicate data
- Also order by the member name.

```python
%sql
WITH full_data AS(
select * from bookings as a
JOIN facilities as b on a.facid = b.facid
JOIN members as c on a.memid = c.memid
WHERE name LIKE '%Tennis Court%')

SELECT DISTINCT name, CONCAT(firstname,' ',surname)
FROM full_data
ORDER BY 2
```

#### Q8: How can you produce a list of bookings on the day of 2012-09-14 which will cost the member (or guest) more than $30? 

- Remember that guests have different costs to members (the listed costs are per half-hour 'slot')
- The guest user's ID is always 0. 

#### Include in your output the name of the facility, the name of the member formatted as a single column, and the cost.

- Order by descending cost, and do not use any subqueries.

```python
%sql
WITH full_data AS(
select * from bookings as a
JOIN facilities as b on a.facid = b.facid
JOIN members as c on a.memid = c.memid),

final_costs AS
(SELECT 
  bookid, 
  name,
  CONCAT(firstname,' ',surname) AS full_name,
  slots,
  membercost, 
  guestcost,
  membercost*slots AS final_m_cost, 
  guestcost*slots AS final_g_cost   
  FROM full_data
WHERE starttime BETWEEN '2012-09-14' AND '2012-09-15')

select
  bookid,
  name,
  full_name,
  final_m_cost,
  final_g_cost
FROM
  final_costs
WHERE
  (final_m_cost > 30 OR final_g_cost > 30)
ORDER BY 4,5


 
```

#### Q9: This time, produce the same result as in Q8, but using a subquery.

```python
%sql
SELECT
bookid,
name,
full_name,
final_m_cost,
final_g_cost
FROM
(select *,
  CONCAT(firstname,' ',surname) AS full_name,
  membercost*slots AS final_m_cost, 
  guestcost*slots AS final_g_cost   
from bookings as a
JOIN facilities as b on a.facid = b.facid
JOIN members as c on a.memid = c.memid
WHERE starttime BETWEEN '2012-09-14' AND '2012-09-15'
AND (membercost*slots > 30
OR guestcost*slots > 30)) as new_table
ORDER BY 4,5
```

#### Q10: Produce a list of facilities with a total revenue less than 1000.
- The output should have facility name and total revenue, sorted by revenue. 
- Remember that there's a different cost for guests and members!

```python
%sql
WITH full_data AS(
select *,
  membercost*slots AS final_m_cost, 
  guestcost*slots AS final_g_cost,
  (membercost*slots) + (guestcost*slots) AS final_booking_cost 
from bookings as a
JOIN facilities as b on a.facid = b.facid
JOIN members as c on a.memid = c.memid)

-- I AM NOT TAKING INTO CONSIDERATION MONTHLY MAINTENANCE OR INITIAL OUTLAY DUE TO A LACK OF A TIME FRAME
SELECT 
name, 
EXTRACT(YEAR FROM starttime), 
ROUND(SUM(final_m_cost)) AS tot_m_cost,
ROUND(SUM(final_g_cost)) AS tot_g_cost,
ROUND(SUM(final_booking_cost)) AS total_revenue
FROM full_data
GROUP BY 1,2
ORDER BY 5
-- BASED ON THIS DATA ONLY THE TABLE TENNIS, SNOOKER, POOL TABLES AND BADMINTON COURT GENERATE 0 REVENUE FROM MEMBERS. ALL OTHER ARE +REV 
```
